```python
import gradio as gr
from transformers import pipeline
from langdetect import detect
import torch
```

```python

# Check if GPU is available and set the device accordingly
device = 0 if torch.cuda.is_available() else -1  # 0 for GPU, -1 for CPU

# Load the translation and summarization models with device argument
semantic_analyzer = pipeline("sentiment-analysis", model="Abdullah104/amazon_sentiment_analysis", tokenizer="Abdullah104/amazon_sentiment_analysis",device=device)

summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

```

```python
def chunk_text(text, max_length=400):  # Increased chunk size
    chunks = []
    for i in range(0, len(text), max_length):
        chunks.append(text[i:i + max_length])
    return chunks

# Function to fetch article and detect language
def fetch_and_process_comment(comment):

    semantics = semantic_analyzer(comment, max_length=300,truncation = True)[0]['label']
    # Chunk the original news
    chunks = chunk_text(comment)

    # Translate and summarize each chunk
    summaries = []

    for chunk in chunks:

        # Adjusting summarization lengths
        summary = summarizer(chunk, max_length=60, min_length=5, do_sample=False)[0]['summary_text']

        summaries.append(summary)

    # Join the results
    summary = ' '.join(summaries)

    return semantics, summary
```

```python
# Create Gradio interface
interface = gr.Interface(
    fn=fetch_and_process_comment,
    inputs=gr.Textbox(label = 'Comment'),
    outputs=[gr.Textbox(label= "sentiments"),gr.Textbox(label= "Summarized Comment")],
    title="Sentiment Analysis and Summarization",
    description="Fetch an article, detect its language, translate to English, and summarize it."
)
```
